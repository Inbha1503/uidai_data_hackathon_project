```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tamilnadu_uidai.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["date"]=pd.to_datetime(df["date"], format="%d-%m-%Y")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['state'],inplace=True)
```

```python
df.sort_values(by='date',inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["total_enrolment"]=(df['age_0_5']+df['age_5_17']+df['age_18_greater'])
```

```python
df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month
```

```python
df['month_name'] = df['date'].dt.month_name()
```

```python
df.head()
```

```python
df.info()
```

```python
df[['age_0_5', 'age_5_17', 'age_18_greater', 'total_enrolment']].describe()
```

```python
age_totals = df[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

age_totals.plot(kind='bar')
plt.title('Age-wise Aadhaar Enrolment Distribution in Tamil Nadu')
plt.xlabel('Age Group')
plt.ylabel('Total Enrolments')
plt.show()
```

```python
monthly_trend = df.groupby('date')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

monthly_trend.plot(figsize=(10,5))
plt.title('Age-wise Aadhaar Enrolment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Enrolments')
plt.show()
```

```python
district_totals = df.groupby('district')['total_enrolment'].sum().sort_values(ascending=False).head(10)

district_totals.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Districts by Aadhaar Enrolment')
plt.xlabel('District')
plt.ylabel('Total Enrolments')
plt.xticks(rotation=45)
plt.show()
```

```python
low_districts = df.groupby('district')['total_enrolment'].sum().sort_values().head(10)

low_districts.plot(kind='bar', figsize=(10,5))
plt.title('Lowest Enrolment Districts')
plt.xlabel('District')
plt.ylabel('Total Enrolments')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_total = df.groupby('date')['total_enrolment'].sum()

monthly_total.plot(figsize=(10,5))
plt.title('Monthly Aadhaar Enrolment Trend')
plt.xlabel('Date')
plt.ylabel('Total Enrolments')
plt.show()
```

```python
monthly_total = df.groupby('date')['total_enrolment'].sum()

mean_val = monthly_total.mean()
std_val = monthly_total.std()

upper_threshold = mean_val + 2 * std_val
lower_threshold = mean_val - 2 * std_val

anomalies = monthly_total[
    (monthly_total > upper_threshold) |
    (monthly_total < lower_threshold)
]

anomalies
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_total)
plt.axhline(upper_threshold, linestyle='--')
plt.axhline(lower_threshold, linestyle='--')
plt.title('Monthly Aadhaar Enrolment with Anomaly Thresholds')
plt.xlabel('Date')
plt.ylabel('Total Enrolments')
plt.show()
```

```python
district_avg = df.groupby('district')['total_enrolment'].mean()

mean_d = district_avg.mean()
std_d = district_avg.std()

high_anomaly_districts = district_avg[district_avg > mean_d + 2*std_d]
low_anomaly_districts = district_avg[district_avg < mean_d - 2*std_d]

high_anomaly_districts, low_anomaly_districts
```

```python
age_means = df[['age_0_5', 'age_5_17', 'age_18_greater']].mean()

age_means
```

```python
district_median = df.groupby('district')['total_enrolment'].median()

low_activity_districts = district_median[district_median < 1]

low_activity_districts
```
